```python
file_path = '/content/gdrive/My Drive/diseases_with_description.csv'
all_chapters = pd.read_csv(file_path)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Initialize CountVectorizer and TfidfVectorizer with English stop words
cv = CountVectorizer(stop_words="english")
cv_tfidf = TfidfVectorizer(stop_words="english")

# Fit and transform using CountVectorizer
X = cv.fit_transform(all_chapters['Description'].tolist())

# Fit and transform using TfidfVectorizer
X_tfidf = cv_tfidf.fit_transform(all_chapters['Description'].tolist())

# Extract feature names from the vectorizers
feature_names_cv = cv.get_feature_names_out()
feature_names_tfidf = cv_tfidf.get_feature_names_out()

# Create DataFrames from the transformed arrays
df_cv = pd.DataFrame(X.toarray(), columns=feature_names_cv)
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=feature_names_tfidf)
```

```python
import numpy as np

# Define a cosine similarity function
cosine = lambda v1, v2: np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

# Define the new text to compare
new_text = ["dizziness loss of balance vomiting tinnitus of hearing in the high frequency range in one ear difficulty focusing your eyes"]

# Transform the new text using CountVectorizer and TfidfVectorizer
new_text_cv = cv.transform(new_text).toarray()[0]
new_text_tfidf = cv_tfidf.transform(new_text).toarray()[0]

# Compare the new text with each chapter in the dataset
for chapter_number in range(all_chapters.shape[0]):
    print(f"This is chapter number: {chapter_number}")
    print(f"Cosine CV:    {cosine(df_cv.iloc[chapter_number], new_text_cv)}")
    print(f"Cosine TFIDF: {cosine(df_tfidf.iloc[chapter_number], new_text_tfidf)}")
```
